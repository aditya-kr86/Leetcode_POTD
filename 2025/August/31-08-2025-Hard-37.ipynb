# Array | Hash Table | Backtracking | Matrix




```python
class Solution:
    def solveSudoku(self, board: list[list[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        self.solve(board)

    def solve(self, board: list[list[str]]) -> bool:
        for i in range(9):
            for j in range(9):
                if board[i][j] == '.':
                    for d in range(1, 10):
                        char_d = str(d)
                        if self.isValid(board, i, j, char_d):
                            board[i][j] = char_d
                            if self.solve(board):
                                return True
                            board[i][j] = '.'  # backtrack
                    return False
        return True

    def isValid(self, board: list[list[str]], row: int, col: int, d: str) -> bool:
        for i in range(9):
            if board[i][col] == d:
                return False
            if board[row][i] == d:
                return False

        start_row = (row // 3) * 3
        start_col = (col // 3) * 3

        for i in range(3):
            for j in range(3):
                if board[start_row + i][start_col + j] == d:
                    return False
        return True
```

```python
sol = Solution()
board = [["5","3",".",".","7",".",".",".","."],
        ["6",".",".","1","9","5",".",".","."],
        [".","9","8",".",".",".",".","6","."],
        ["8",".",".",".","6",".",".",".","3"],
        ["4",".",".","8",".","3",".",".","1"],
        ["7",".",".",".","2",".",".",".","6"],
        [".","6",".",".",".",".","2","8","."],
        [".",".",".","4","1","9",".",".","5"],
        [".",".",".",".","8",".",".","7","9"]]
sol.solveSudoku(board=board)
print(board)
```
